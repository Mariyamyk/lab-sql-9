```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
```

```python
import urllib.parse
urllib.parse.quote_plus("")
```

```python
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost:3306/sakila'
engine = create_engine(connection_string)
engine
```

```python

#Check the number of rentals for each customer for May
query_may='''
            SELECT customer_id, count(rental_id) as rent_may 
            FROM rentals_may 
            GROUP by customer_id;'''
            
rentals_may_count= pd.read_sql_query(query_may, engine)
rentals_may_count

```

```python
#Check the number of rentals for each customer for June

query_june='''
            SELECT customer_id, count(rental_id) as rent_june 
            FROM rentals_june 
            GROUP by customer_id;'''
            
rentals_june_count= pd.read_sql_query(query_june, engine)
rentals_june_count
```

```python
#Write a function that checks if customer borrowed more or less films in the month of June as compared to May.
rents=pd.merge(rentals_may_count,rentals_june_count, on=['customer_id'], how='right' )
rents
```

```python
rents['rent_may'] = rents['rent_may'].fillna(0)
rents
```

```python
rents['rent_june'] = rents['rent_june'].fillna(0)
rents
```

```python
rents['Rents_delta']=rents['rent_june']-rents['rent_may']
rents
```
